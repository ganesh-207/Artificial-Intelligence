## WordNet WordNet is a lexical database of English that groups words into synsets(set of synonyms) and defines relationships between them such as:
  Synonym (similar meaning)
Antonym (opposites)
Hypernym/Hyponym (general–specific hierarchy)
Meronym/Holonym (part–whole relationship)

It becomes necessary (or useful) in tasks where you need semantic meaning of words, not just raw text.

## Hypernyms, Hyponyms, Meronyms
## Hypernym: A hypernym is a word that serves as a general category or umbrella term for a group of related words, known as hyponyms, which are more specific
## Hyponym: A hyponym is a specific example falling under hypernym
## Meronym: Meronyms are parts that make up the whole item.
## Holynyms: A holonym is a word that refers to a whole entity of which a given word is a part, such as "hat" being a holonym for "brim" or "crown

  Hypernym and Hyponym Example:<br>
    Animal<br>
      Mammal<br>
          Dog (Labrador, Poodle)<br>
          Cat (Siamese, Persian)<br>
      Bird<br>
          Night Birds(Owl,Bat)<br>
          Morning Birds(Parrot, Eagle)<br>
    Animal is a hypernym
    Mammal and Bird are Hyponym for Animal hypernym
     Dog,Cat are Hyponyms for Mammal hypernym and Night Birds, Mornings are Hyponym for Bird hypernym

  Meronym Example:<br>
  Car<br>
  Engine (Pistons, Crankshaft)<br>
  Wheels (Tires, Rims)<br>
  Body (Doors, Windows)<br>

  Car is Holonym and Engine, Wheels, Body are Meronym

## Measures of Semantic Similarity
It’s the degree to which two words or concepts are related in meaning.

Lexical similarity → based on characters/letters (e.g., “cat” vs “bat”).

Semantic similarity → based on meaning (e.g., “car” vs “automobile”).

## Most common similarities:
Path Similarity:

Based on the shortest path (number of edges) between synsets in the WordNet hierarchy.
Range: 0 to 1 (1 if they are the same synset).
Example: dog vs cat.<br>
Wu–Palmer Similarity (wup_similarity)

Based on the depth of the concepts and their Least Common Subsumer (LCS) (lowest shared ancestor).
Range: 0 to 1.
Higher if two concepts share a deeper ancestor.

## Word Sense Disambighution

```python
from nltk.wsd import lesk
```

```python
sentence1 = "He deposited the cheque at the bank"
sentence2 = "He sat on the bank of the river."
```

```python
from nltk.tokenize import word_tokenize
words1 = word_tokenize(sentence1)
words2 = word_tokenize(sentence2)

wsd1 = lesk(words1,'bank')
wsd2 = lesk(words2,'bank')

print(sentence1)
print(f"Disambiguated to {wsd1.definition()}")

print("================",sentence2,sep='\n')
print(f"Disambigued Sence: {wsd2.definition()}")
```

## There are some disadvantages to simplified version of lesk as it uses overlap of dictionary glosses. We can instead use pywsd library

```python
from pywsd.lesk import simple_lesk

sentence1 = "He deposited the cheque at the bank"
sentence2 = "He sat on the bank of the river"

sl_wsd1 = simple_lesk(sentence1,'bank')
sl_wsd2 = simple_lesk(sentence2,'bank')

print(sentence1)
print(f"Disambiguated to {sl_wsd1.definition()}")

print("==================",sentence2,sep='\n')
print(f"Disambigued Sense: {sl_wsd2.definition()}")
```

```python
import gensim
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import gutenberg
```

```python
text = gutenberg.raw("shakespeare-hamlet.txt")
sentences = sent_tokenize(text)

hamlet = [word_tokenize(sent.lower()) for sent in sentences]
```

```python
hamlet[:50]
```

## Using CBOW Model

## We will be using the CBOW architecture:

  min_count=1: Includes words that appear at least once.<br>
  vector_size=100: Generates word vectors of 100 dimensions.<br>
  window=5: Considers a context window of 5 words before and after the target word.<br>
  sg=0: Uses CBOW model (default setting).

```python
model1 = gensim.models.Word2Vec(hamlet, min_count=1,
                                vector_size=100, window=5)
# model1 = gensim.models.Word2Vec(hamlet, min_count=1,
                                  # vector_size=100,window=5,sg=0)
```

## Using Skip-Gram Model

## We will be using the Skip-Gram architecture for this model.

  min_count=1: Includes words that appear at least once.<br>
  vector_size=100: Generates word vectors of 100 dimensions.<br>
  window=5: Considers a context window of 5 words.<br>
  sg=1: Enables the Skip-Gram model (predicts context words given a target word).

```python
model2 = gensim.models.Word2Vec(hamlet, min_count=1, vector_size=100, window=5, sg=1)
```

## Evaluating Word Similarities

## We will compute the cosine similarity between word vectors to assess semantic similarity. Cosine similarity values range from -1 (opposite) to 1 (very similar), showing how closely related two words are in terms of meaning.

## model.wv.similarity(word1, word2): Computes the cosine similarity between word1 and word2 based on the trained model.

```python
print("Cosine similarity between 'william' " +
      "and 'shakespeare' - CBOW : ",
      model1.wv.similarity('william','shakespeare'))

print("Cosine similarity between 'actus' " +
      "and 'primus' - Skip-Gram : ",
      model2.wv.similarity('actus', 'primus'))
```

```python
# part 1: Characters that interact in Hamlet
print("Cosine similarity between 'hamlet' and 'horatio' - CBOW: ",
      model1.wv.similarity('hamlet','horatio'))

print("Cosine similarity between 'king' and 'queen' - Skip-Gram: ",
      model2.wv.similarity('king','queen'))
```

## Glove Model

```python
import gensim.downloader as api

# load pre-trained Glove model (50-dimensional vectors, trained on Wikipedia + Gigaword)
glove = api.load("glove-wiki-gigaword-50")  # When loaded, glove behaves like a dictionary: keys = words,values = vectors.

# Similarity between related words-computes cosine similarity between the two word vectors.
print("Similarity between king and queen:",glove.similarity("king","queen"))
print("Similarity between paris and france:",glove.similarity("paris","france"))

# Similarity between unrelated words-computes cosine similarity between the two word vectors.
print("Similarity between king and banana:",glove.similarity("king","banana"))
print("Similarity between computer and love:",glove.similarity("computer","love"))

# Famous analogy: king - man + woman = queen
result = glove.most_similar(positive=["king","woman"],negative=["man"],topn=1)
print("Analogy king - man + woman ->",result)
```
